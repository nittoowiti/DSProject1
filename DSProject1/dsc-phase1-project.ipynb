```python
import pandas as pd
merged_df = pd.merge(df1, df2, on='id', how='inner')

# Display the merged dataframe
merged_df.head()
```

```python
import pandas as pd

# Convert box_office, production_budget, domestic_gross, and worldwide_gross columns to numeric
merged_df['box_office'] = pd.to_numeric(merged_df['box_office'].replace('[\$,]', '', regex=True), errors='coerce')
merged_df['production_budget'] = pd.to_numeric(merged_df['production_budget'].replace('[\$,]', '', regex=True), errors='coerce')
merged_df['domestic_gross'] = pd.to_numeric(merged_df['domestic_gross'].replace('[\$,]', '', regex=True), errors='coerce')
merged_df['worldwide_gross'] = pd.to_numeric(merged_df['worldwide_gross'].replace('[\$,]', '', regex=True), errors='coerce')

# Calculate ROI and add a new column
merged_df['roi'] = (merged_df['worldwide_gross'] - merged_df['production_budget']) / merged_df['production_budget'] * 100

# Print the updated DataFrame
merged_df.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Calculate the mean ROI for each studio
average_roi_per_studio = merged_df.groupby('studio')['roi'].mean()

# Select the top studios by ROI
top_studios = average_roi_per_studio.sort_values(ascending=False).head(10)

# Plotting the top studios by ROI
plt.figure(figsize=(12, 6))
top_studios.sort_values().plot(kind='barh', color='purple')
plt.xlabel('Average ROI (%)')
plt.ylabel('Studio')
plt.title('Top Studios by Average ROI')
plt.grid(axis='x')

# Display the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plotting box plot for ROI by studio
plt.figure(figsize=(16, 8))
sns.boxplot(x='roi', y='studio', data=merged_df, palette='viridis')
plt.xlabel('ROI (%)')
plt.ylabel('Studio')
plt.title('ROI Distribution by Studio')
plt.grid(True)

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Select the top ten movies by ROI
top_10_movies = merged_df.sort_values(by='roi', ascending=False).head(10)

# Plotting the top ten movies by ROI
plt.figure(figsize=(12, 6))
plt.barh(top_10_movies['movie'], top_10_movies['roi'], color='green')
plt.xlabel('ROI (%)')
plt.ylabel('Movie')
plt.title('Top 10 Movies by ROI')
plt.grid(axis='x')

# Display the plot
plt.show()
```

```python
#merge tmdb.movies.csv & tn.movies_budget.csv

# Load the datasets
budgets = df2
movies = df3

# 1. Handle missing values
budgets.dropna(inplace=True)
movies.dropna(inplace=True)

# 2. Convert date columns
budgets['release_date'] = pd.to_datetime(budgets['release_date'])
movies['release_date'] = pd.to_datetime(movies['release_date'])

# 3. Convert budget and gross columns
budgets['production_budget'] = budgets['production_budget'].replace('[\$,]', '', regex=True).astype(float)
budgets['domestic_gross'] = budgets['domestic_gross'].replace('[\$,]', '', regex=True).astype(float)
budgets['worldwide_gross'] = budgets['worldwide_gross'].replace('[\$,]', '', regex=True).astype(float)

# 4. Merge datasets
merged_data = pd.merge(budgets, movies, how='inner', left_on='movie', right_on='original_title')

# 5. Explore and analyze
# For example, you can explore correlations, trends, or summary statistics.

# Display the first few rows of the merged dataset
merged_data.head(5)

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Select numerical columns for distribution plots
numerical_columns = ['production_budget', 'domestic_gross', 'worldwide_gross', 'popularity', 'vote_average', 'vote_count']

# Set up the subplot grid
fig, axes = plt.subplots(nrows=len(numerical_columns), ncols=1, figsize=(10, 12), sharex=False)

# Plot distribution plots for each numerical column
for i, column in enumerate(numerical_columns):
    sns.histplot(merged_data[column], kde=True, ax=axes[i], color='skyblue', bins=30)
    axes[i].set_title(f'Distribution of {column}')
    axes[i].set_xlabel(column)

# Adjust layout
plt.tight_layout()

# Display the plots
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Convert 'release_date_x' to datetime format
merged_data['release_date_x'] = pd.to_datetime(merged_data['release_date_x'])

# Filter data for movies released from the year 2005 onwards
filtered_data = merged_data[merged_data['release_date_x'].dt.year >= 2005]

# Extract year and month from release date
filtered_data['release_year'] = filtered_data['release_date_x'].dt.year
filtered_data['release_month'] = filtered_data['release_date_x'].dt.month_name()

# Count the number of movies released each month
monthly_movie_counts = filtered_data.groupby(['release_year', 'release_month']).size().unstack().fillna(0)

# Plot the bar plot
plt.figure(figsize=(15, 8))
monthly_movie_counts.plot(kind='bar', stacked=True, colormap='viridis')
plt.title('Monthly Distribution of Movies Released (2005 Onwards)')
plt.xlabel('Release Year')
plt.ylabel('Number of Movies')
plt.legend(title='Month', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Create a bar plot for original language vs. popularity
plt.figure(figsize=(12, 6))
sns.barplot(x='original_language', y='popularity', data=merged_data, ci=None, palette='viridis')
plt.title('Original Language vs. Popularity')
plt.xlabel('Original Language')
plt.ylabel('Popularity')
plt.show()
```

```python
import matplotlib.pyplot as plt

popularity = merged_data['popularity']
domestic_gross = merged_data['domestic_gross']
worldwide_gross = merged_data['worldwide_gross']

# Sort the data by popularity for better visualization
sorted_indices = popularity.argsort()
popularity = popularity[sorted_indices]
domestic_gross = domestic_gross[sorted_indices]
worldwide_gross = worldwide_gross[sorted_indices]

# Create a line plot
plt.figure(figsize=(12, 8))
plt.plot(popularity, domestic_gross, label='Domestic Gross', marker='o', linestyle='-')
plt.plot(popularity, worldwide_gross, label='Worldwide Gross', marker='o', linestyle='-')

# Add labels and title
plt.title('Popularity vs. Gross Revenue')
plt.xlabel('Popularity')
plt.ylabel('Gross Revenue')
plt.legend()
plt.grid(True)
plt.show()
```
